### upload

clash 代理

access token

```python
from huggingface_hub import HfApi
import os

os.environ["HTTP_PROXY"] = "http://127.0.0.1:7890"
os.environ["HTTPS_PROXY"] = "http://127.0.0.1:7890"

# 配置你的信息
TOKEN = "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"
REPO_ID = "Cyfef/UniMesh_base"
BRANCH = "glbs_4"

api = HfApi(token=TOKEN)

print(f"正在检查/创建分支: {BRANCH}...")
api.create_branch(
    repo_id=REPO_ID, 
    repo_type="dataset", 
    branch=BRANCH,
    exist_ok=True # 如果分支已存在，不会报错
)
print(f"分支 {BRANCH} 已准备就绪。")

print(f"开始上传文件到 {REPO_ID} 的 {BRANCH} 分支...")

try:
    # upload_folder 会自动处理 LFS 和大于 5GB 的文件
    # multi_commits=True 会将大文件切片上传，避开所有 5GB 限制报错
    api.upload_folder(
        folder_path="./Diffurank/valid/UniMesh_base", 
        repo_id=REPO_ID,
        repo_type="dataset",
        revision=BRANCH,
        #multi_commits=True, 
        ignore_patterns=[".git/*"]
    )
    print("上传成功！请前往 Hugging Face 页面查看。")
except Exception as e:
    print(f"上传失败: {e}")
```
