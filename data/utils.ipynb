# 解压和删除无gt的物体 

```python
import os
import zipfile
import shutil

def unzip_and_delete_objs(folder_path):
    
    for file_name in os.listdir(folder_path):
        # 筛选压缩包
        if file_name.endswith('.zip'):
            # 创建解压目录
            file_abs_path = os.path.join(folder_path,file_name)
            extract_dir = os.path.splitext(file_abs_path)[0]
            
            try:
                print(f"正在解压: {file_name}...")
                with zipfile.ZipFile(file_abs_path, 'r') as zip_ref:
                    # 解压到指定目录
                    zip_ref.extractall(extract_dir)
                
                # 解压成功后删除原文件
                os.remove(file_abs_path)
                print(f"解压完成并已删除: {file_name}")
                
            except Exception as e:
                print(f"处理 {file_name} 时出错: {e}")
            
            # 筛选带有gt caption
            for obj_name in os.listdir(os.path.join(extract_dir,)):
                if obj_name not in keys_list:
                    shutil.rmtree(os.path.join(extract_dir,,obj_name))
```

```python
data_dir=""
unzip_and_delete_objs(data_dir)
```

# 筛选latent codes

```python
objs_path=""
objs_list=os.listdir(objs_path)
```

```python
import os
import zipfile

def unzip_and_delete_latent_codes(folder_path):
    
    for file_name in os.listdir(folder_path):
        # 筛选压缩包
        if file_name.endswith('.zip'):
            # 创建解压目录
            file_abs_path = os.path.join(folder_path,file_name)
            extract_dir = os.path.splitext(file_abs_path)[0]
            
            try:
                print(f"正在解压: {file_name}...")
                with zipfile.ZipFile(file_abs_path, 'r') as zip_ref:
                    # 解压到指定目录
                    zip_ref.extractall(extract_dir)
                
                # 解压成功后删除原文件
                os.remove(file_abs_path)
                print(f"解压完成并已删除: {file_name}")
                
            except Exception as e:
                print(f"处理 {file_name} 时出错: {e}")
            
            # 筛选带有gt caption
            for file in os.listdir(os.path.join(extract_dir,)):
                obj_name=os.path.splitext(file)[0]
                if obj_name not in objs_list:
                    os.remove(os.path.join(extract_dir,,file))
```

```python
dir_path=""
unzip_and_delete_latent_codes(dir_path)
```

# 根据 latent codes 筛选物体

```python
latent_codes_path=""
lat_codes_list=[os.path.splitext(f_name)[0] for f_name in os.listdir(latent_codes_path)]
```

```python
import os
import shutil

def delete_objs_based_on_codes(objects_dir):
    for obj_name in os.listdir(objects_dir):
        if obj_name not in lat_codes_list:
            shutil.rmtree(os.path.join(objects_dir,obj_name))
```

```python
objects_dir=""
delete_objs_based_on_codes(objects_dir)
```
