```python
#!/bin/bash
#SBATCH --job-name=unimesh1
#SBATCH --output=/data/group/zhaolab/project/UniMesh/UniMesh_und/Diffurank/shap-e/latent.txt
#SBATCH --ntasks=1
#SBATCH --mem=200G
#SBATCH --gres=gpu:1
#SBATCH --partition=week

echo "Job started on:"
date

# 加载 Xvfb 模块（使用绝对路径或 module 命令）
# 方法一：使用 module 命令（需确保计算节点支持）
module load Xvfb/21.1.3   # 或 Xvfb/1.20.9

# 方法二：如果 module 不可用，直接指定 Xvfb 绝对路径（推荐）
# 先通过 `module show Xvfb/21.1.3` 获取实际路径，例如：
# XVFB=/usr/local/easybuild/software/Xvfb/21.1.3-GCCcore-11.3.0/bin/Xvfb
# 然后启动：
# $XVFB :99 -screen 0 1024x768x24 &

# 启动 Xvfb（假设方法一成功）
Xvfb :99 -screen 0 1024x768x24 &
XVFB_PID=$!
export DISPLAY=:99

# 运行 Python 脚本

export LD_LIBRARY_PATH=/data/group/zhaolab/project/miniconda/envs/diffurank/lib:$LD_LIBRARY_PATH

python extract_latent.py --uid_path "../UniMesh_base/object_path.pkl" --save_name "../UniMesh_base/extracted_shapE_latent"

# 清理
kill $XVFB_PID

echo "Job finished on:"
date
```
